Create a pivot table showing average income by Department and JobRole.



```python
plt.figure(figsize=(8, 5))
sns.histplot(data=data, x="Age", hue="Attrition", kde=True, multiple="stack")
plt.title("Age Distribution by Attrition")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
average_income=pd.pivot_table(data,index="Department",columns="JobRole",values="MonthlyIncome",aggfunc="median")
```

```python
plt.figure(figsize=(12, 8))
corr = data.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
